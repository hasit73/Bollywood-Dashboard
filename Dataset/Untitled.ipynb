```python
d1 = d[['imdbId',"movieBudget","movieProfit","genre"]].set_index(['imdbId',"movieBudget","movieProfit"]).apply(lambda x: x.str.split('|').explode()).reset_index()
```

```python
d1 = d1.dropna(subset=["genre"])
```

```python
d1["genre"] = d1["genre"].apply(lambda x: x.lower().strip())
d1
```

```python
newdata = d1.merge(d,how="inner",on="imdbId")[["imdbId","Target age group","genre_x"]]
newdata
```

```python
crossdata = pd.crosstab(index=newdata["Target age group"],columns=newdata["genre_x"])
```

```python
def calculatePercentage(genre,agegroup,data):
    t1 = data[data["genre_x"]==genre]
    t2 = t1[t1["Target age group"]==agegroup]["imdbId"].count()
    return str(round(100*t2/t1["imdbId"].count()))+" %"
calculatePercentage("action","26-35",newdata)
```

```python
new2 = crossdata.copy()
```

```python
new2.loc["14-25","action"]
```

```python
for i in crossdata.columns:
    for j in crossdata.index:
        new2.loc[j,i] = calculatePercentage(i,j,newdata)
```
